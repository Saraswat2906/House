<H1> Load the data

```python
data=pd.read_csv('Raw_Housing_Prices.csv')
data = data.fillna(10000)  # Replace 'value' with the value you want to use for filling NaN.

data
```

```python
print(data.describe())
```

```python
cols_to_drop=['Waterfront View','No of Times Visited','Condition of the House','Overall Grade','Renovated Year','Date House was Sold']
data=data.drop(cols_to_drop, axis=1)
```

```python
sns.heatmap(data.corr(),annot = True)
```

#Check missing Data

```python
data.isna().sum()
```

#Data Normalisation

```python
data = data.iloc[:,1:]
data_norm = (data - data.mean()) / data.std()
data_norm.head()
```

```python
y_mean = data['Sale Price'].mean()
y_std = data['Sale Price'].std()

def convert_label_value(pred):
    return int(pred * y_std + y_mean)
```

```python
X = data_norm.iloc[:, 1:14]
X.head()
```

```python
Y = data_norm.iloc[:, 0]
Y.head()
```

```python
X_arr = X.values
Y_arr = Y.values

print('X_arr shape: ', X_arr.shape)
print('Y_arr shape: ', Y_arr.shape)
```

```python


# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_arr, Y_arr, test_size=0.2, random_state=42)

# Build the neural network model
model = Sequential()

model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Define early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=10)

# Train the model
history = model.fit(
    X_train, Y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_test, Y_test),
    callbacks=[early_stopping]
)

# Evaluate the model on the test set
loss = model.evaluate(X_test, Y_test)
print(f'Mean Squared Error on Test Set: {loss}')

# Save the trained model
model.save('house_price_model.h5')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Assuming 'model' is your trained neural network model

# Predictions on the test set before training
preds_on_untrained = model.predict(X_test)

# Train the model (if not already trained)

# Predictions on the test set after training
preds_on_trained = model.predict(X_test)

# Function to compare predictions
def compare_predictions(preds1, preds2, actual):
    plt.figure(figsize=(10, 6))
    plt.scatter(actual, preds1, label='Predictions Before Training', alpha=0.5)
    plt.scatter(actual, preds2, label='Predictions After Training', alpha=0.5)
    plt.plot([min(actual), max(actual)], [min(actual), max(actual)], '--', color='red', label='Ideal')
    plt.title('Comparison of Predictions Before and After Training')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.legend()
    plt.show()

# Call the function to compare predictions
compare_predictions(preds_on_untrained, preds_on_trained, Y_test)
```

```python
# Assuming 'model' is your trained neural network model

# Predictions on the test set before training
preds_on_untrained = model.predict(X_test)

# Predictions on the test set after training
preds_on_trained = model.predict(X_test)

# Convert predicted values to the original scale
price_on_untrained = [convert_label_value(y) for y in preds_on_untrained]
price_on_trained = [convert_label_value(y) for y in preds_on_trained]
price_y_test = [convert_label_value(y) for y in Y_test]

# Function to compare predictions
def compare_predictions(preds1, preds2, actual):
    plt.figure(figsize=(10, 6))
    plt.scatter(actual, preds1, label='Predictions Before Training', alpha=0.5)
    plt.scatter(actual, preds2, label='Predictions After Training', alpha=0.5)
    plt.plot([min(actual), max(actual)], [min(actual), max(actual)], '--', color='red', label='Ideal')
    plt.title('Comparison of Predictions Before and After Training')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.legend()
    plt.show()

# Call the function to compare predictions
compare_predictions(price_on_untrained, price_on_trained, price_y_test)
```
